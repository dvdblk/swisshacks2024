```python
import pandas as pd

DATA_PATH = "../../data/jb-swisshacks2024/"
is_fake_data = pd.read_csv(DATA_PATH + "fake_audio_classification_results.csv")
is_factually_correct = pd.read_csv(DATA_PATH + "fact_checks_v2.csv")
```

```python
# Extract the FINAL_LABEL column and rename it to is_fake
is_fake_data['is_fake'] = is_fake_data['FINAL_LABEL'].apply(lambda x: True if x == 1 else False)

# keep only is_fake and ID columns
is_fake_data = is_fake_data[['ID', 'is_fake']].rename(columns={'ID': 'rec_id'})
is_fake_data
```

```python
df_fake_and_facts = pd.merge(is_fake_data, is_factually_correct, on="rec_id")

# transform is_factually_correct from 'no' to False and 'yes' to True
df_fake_and_facts['is_factually_correct'] = df_fake_and_facts['is_factually_correct'].apply(lambda x: True if x == 'yes' else False)

# keep only the rec_id, is_fake and is_factually_correct columns
df_fake_and_facts = df_fake_and_facts[['rec_id', 'is_fake', 'is_factually_correct']]
df_fake_and_facts
```

```python
# export the final dataframe
df_fake_and_facts.to_csv(DATA_PATH + "prefinal_is_fake_and_facts.csv", index=False)
```
